```python
df = pd.read_csv(WORKSPACE_ROOT/"data/proteomics/epi300_proteomics.protein.tsv", sep="\t")
COLUMNS = [
    "Protein","Description",
    "Protein Length",
    "Combined Total Peptides",
    "Combined Spectral Count",
    "Combined Unique Spectral Count",
    "Combined Total Spectral Count",
    "76_Pic1011_4_F4_7G_100ng Total Spectral Count",
    "76_Pic1011_4_F4_7G_100ng Intensity",
]
new_names = [x.replace("Combined ", "").strip() for x in COLUMNS]
df_controls = df[df.Protein.str.startswith("sp")][COLUMNS]
df_controls.columns = new_names
df_controls
```

```python
df_controls.to_csv("./epi300_proteomics.controls.csv")
```
